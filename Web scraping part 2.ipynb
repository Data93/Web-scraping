```python
#import package yang dibutuhkan
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Agar tampilan content menjadi lebih rapi, kita akan menggunakan BeautifulSoup constructor dan bantuan HTML parser.
soup = BeautifulSoup(getRequest.text, 'html.parser')
print(soup.prettify())
```

```python
# Mendefinisikan tempat yang menampung semua url halaman
url_pages = []

# Mendefinisikan halaman pertama adalah 3
page = 3
init_page = f"https://books.toscrape.com/catalogue/page-{page}.html"

# Looping pengecekan status halaman
while requests.get(init_page).status_code == 200:
  url_pages.append(init_page)
  page += 3
  init_page = f"https://books.toscrape.com/catalogue/page-{page}.html"
```

```python
#Membuat suatu fungsi untuk mendapatkan/mengakses/mengunduh webpage
def getAndParseURL(url):
  getRequest = requests.get(url)
  soup = BeautifulSoup(getRequest.text, 'html.parser')
  return soup
```

```python
# Misalkan webpage yang akan kita unduh "https://books.toscrape.com/catalogue/page-3.html"
# Cek fungsi getAndParseURL
url = "https://books.toscrape.com/catalogue/page-3.html"
soup = getAndParseURL(url)
soup
```

```python
# Langkah 2a. Mendapatkan url satu buku
# Letak url buku : berada di dalam tag article dengan class "product_pod" --> <h3> --> a

# soup.find("article", class_ = "product_pod")
# soup.find("article", class_ = "product_pod").h3
# soup.find("article", class_ = "product_pod").h3.a
url = soup.find("article", class_ = "product_pod").h3.a.get('href') # .find() digunakan untuk mencari tag dan class

# karena hasil yang didapatkan belum sepenuhnya url dari buku yanng dimaksud, maka kita bentuk bagian-bagian url yang kurang
url = f"https://books.toscrape.com/catalogue/{url}"
url
```

```python
# Langkah 2b. Mendapatkan semua url buku di satu halaman

# soup.findAll("article", class_ = "product_pod") # Mendapatkan semua tag article dan class "product_pod" dalam satu halaman
url_books = [f"https://books.toscrape.com/catalogue/{x.h3.a.get('href')}" for x in soup.findAll("article", class_ = "product_pod")]
url_books
```

```python
def getBooksURLs(url):
  soup = getAndParseURL(url)
  url_books = [f"https://books.toscrape.com/catalogue/{x.h3.a.get('href')}" for x in soup.findAll("article", class_ = "product_pod")]
  return url_books
```

```python
# Cek fungsi getBooksURLs()
url = "https://books.toscrape.com/catalogue/page-3.html"
getBooksURLs(url)
```

```python
#Mendapatkan semua url buku pada semua halaman
all_urlBooks = []
for url in url_pages:
  all_urlBooks.extend(getBooksURLs(url))
```

```python
def getBookContents(book_url):
  soup = getAndParseURL(book_url)
  title = soup.find("div", class_ = "col-sm-6 product_main").h1.text
  price = soup.find("p", class_ = "price_color").text
  available = soup.find("p", class_ = "instock availability").text
  available = available.strip()
  url_image = soup.find("article", class_ = "product_page").img.get('src')
  url_image = "https://books.toscrape.com/" + '/'.join(url_image.split('/')[2:])
  rating = soup.find("p", class_ = re.compile("star-rating")).get("class")[1]
  category = soup.find("ul", class_ = "breadcrumb").findAll("li")[2].a.text
  description = soup.find("article", class_ = "product_page").findAll("p")[-1].text
  return {'title': title, 'book_url': book_url, 'price': price, 'available': available, 'url_image': url_image, 'rating': rating, 'category': category, 'description': description}
```

```python
# Cek fungsi getBookContents(), misalnya book_url adalah "https://books.toscrape.com/catalogue/slow-states-of-collapse-poems_960/index.html"
getBookContents("https://books.toscrape.com/catalogue/slow-states-of-collapse-poems_960/index.html")
```

```python
book_collections = pd.DataFrame()
for book_url in all_urlBooks:
  per_book = getBookContents(book_url)
  print(per_book)
  book_collections = book_collections.append(per_book, ignore_index=True)
```
